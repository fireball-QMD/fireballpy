```python
ETOT=[]
F=[]
time=[]
for i in range(opt.get_number_of_steps()+1):
    ETOT.append(traj[i].get_total_energy())
    time.append(i*opt.dt)
    f=traj[i].get_forces()
    m=np.sum(np.linalg.norm(f, axis=1))
    F.append(m)
```
